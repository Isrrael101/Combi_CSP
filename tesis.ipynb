```python
import pandas as pd
import datetime

df_1 = pd.DataFrame(index=df.index)
# Convertir las columnas Year, Month, Day, Hour y Minute a una columna de fecha y hora
df_1['fecha_y_hora'] = pd.to_datetime(df[['Year', 'Month', 'Day', 'Hour', 'Minute']])

# Obtener el primer día del año del registro más temprano en df
primer_dia_del_ano = df_1['fecha_y_hora'].min().replace(hour=0, minute=0, second=0)

# Calcular la hora del año utilizando el primer día del año obtenido
df_1['hora_del_año'] = ((df_1['fecha_y_hora'] - primer_dia_del_ano).dt.total_seconds() / 3600)
df_1
```

```python
import CombiCSP.SolarGeometry as SG
H_msnm = 3706 # Altura de Oruro
df_1['Tiempo_solar'] = SG.tsol(df_1['hora_del_año']) # Tiempo solar en horas
df_1['Ángulo_solar'] = SG.W(df_1['hora_del_año']) # Ángulo horario solar en grados
df_1['elevación_solar'] = SG.ele(df_1['hora_del_año']) # elevación solar en radianes
df_1['Ángulo_zenital'] = SG.z(df_1['hora_del_año'])# Ángulo zenital solar en radianes
df_1['Ángulo_azimutal'] = SG.azim(df_1['hora_del_año']) # Ángulo azimutal solar en radianes
df_1['Ángulo_incidencia'] = SG.thetai(df_1['hora_del_año']) # Ángulo de incidencia solar en radianes
df_1['ISE'] = SG.I0(df_1['hora_del_año']) # Irradiancia solar extraterrestre en W/m^2
df_1['Estado'] = ''
df_1['Masa_Aire'] = SG.air_mass(df_1['hora_del_año']) # Masa de aire (sin especificar el método)
# Agrega una columna 'Estado' que contendrá el texto "Valor negativo" cuando se detecten valores negativos en 'Masa_Aire'
df_1['Estado'] = ''
df_1.loc[df_1['Masa_Aire'] < 0, 'Estado'] = 'Valor negativo'
df_1['Masa_Aire_wiki'] = SG.air_mass(df_1['hora_del_año'], method='wiki') # Masa de aire usando el método 'wiki'
df_1['Masa_Aire_Kasten'] = SG.air_mass(df_1['hora_del_año'], method='Kasten')
df_1['Masa_Aire_Kasten-Young'] = SG.air_mass(df_1['hora_del_año'], method='Kasten-Young')
df_1['Masa_Aire_Schoenberg'] = SG.air_mass(df_1['hora_del_año'], method='Schoenberg')
df_1['Irr_Dir'] = SG.Ib(df_1['hora_del_año'])
alt = 3706
df_1['Irr_Dir_alt'] = SG.Ib2(alt, (df_1['hora_del_año']))

df_1
```
