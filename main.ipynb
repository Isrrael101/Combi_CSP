# Llevar de datetime a horas del año

```python
import pandas as pd
import numpy as np
from datetime import datetime

# Paso 1: Obtener la fecha actual
fecha_actual = datetime.now()

# Paso 2: Calcular la diferencia en horas desde el primer día del año
primer_dia_del_ano = datetime(fecha_actual.year, 1, 1)
diferencia = fecha_actual - primer_dia_del_ano
diferencia_en_horas = diferencia.days * 24 + diferencia.seconds // 3600

# Paso 3: Crear un rango de horas del año
HOYS_DEFAULT = np.arange(1, 8761, 1)

# Paso 4: Mostrar el resultado
print("Fecha actual:", fecha_actual)
print("Horas desde el primer día del año:", diferencia_en_horas)
print("Rango de horas del año:", HOYS_DEFAULT)
```
